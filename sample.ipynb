```python
from nba_api.stats.static import players

def fetch_players():
    all_players = players.get_players()
    players_data = []
    for player in all_players:
        
        players_data.append({
            'player_id': player['id'],
            'player_name': player['full_name'],
            'is_active': player['is_active']
        })
    return players_data
```

```python
from nba_api.stats.static import teams

def fetch_teams():
    all_teams = teams.get_teams()
    teams_data = []
    for team in all_teams:
        # print(team)
        # break
        teams_data.append({
            'team_id': team['id'],
            'team_name': team['full_name'],
            'abbreviation': team['abbreviation'],
            'city': team['city']
            # 'conference': team['conference'],
            # 'division': team['division']
        })
    return teams_data
```

```python
from nba_api.stats.endpoints import leaguegamefinder

def fetch_games(season='2021-22'):
    game_finder = leaguegamefinder.LeagueGameFinder(season_nullable=season)
    games_df = game_finder.get_data_frames()[0]
    games_data = games_df[['GAME_ID', 'GAME_DATE', 'TEAM_ID', 'MATCHUP', 'PTS']].to_dict(orient='records')
    return games_data
```

```python
from nba_api.stats.endpoints import playergamelog

def fetch_player_game_stats(player_id, season='2021-22'):
    game_log = playergamelog.PlayerGameLog(player_id=player_id, season=season)
    game_stats_df = game_log.get_data_frames()[0]
    # print(game_stats_df.columns)
    # return
    game_stats = game_stats_df[['Game_ID', 'Player_ID', 'PTS', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'FG_PCT']].to_dict(orient='records')
    return game_stats
```

```python
from nba_api.stats.endpoints import boxscoreadvancedv2

def fetch_advanced_stats(game_id):
    advanced_stats = boxscoreadvancedv2.BoxScoreAdvancedV2(game_id=game_id)
    advanced_df = advanced_stats.get_data_frames()[0]
    # print(advanced_df.columns)
    advanced_data = advanced_df[['GAME_ID', 'PLAYER_ID', 'TS_PCT', 'EFG_PCT', 'USG_PCT', 'OREB_PCT', 'OFF_RATING', 'OFF_RATING']]
    return advanced_data
```

```python
from nba_api.stats.endpoints import playbyplayv2

def fetch_play_by_play(game_id):
    play_by_play = playbyplayv2.PlayByPlayV2(game_id=game_id)
    play_by_play_df = play_by_play.get_data_frames()[0]
    print(play_by_play_df.columns)
    play_data = play_by_play_df[['GAME_ID', 'PERIOD', 'PCTIMESTRING', 'HOMEDESCRIPTION', 'VISITORDESCRIPTION']]
    return play_data
```

```python
import psycopg2

# Database connection parameters
DB_NAME = 'nba_db'
DB_USER = 'nba_user'
DB_PASSWORD = 'nba_pass'
DB_HOST = 'localhost'
DB_PORT = '5432'

# Function to create tables in PostgreSQL


def create_tables():


    commands = (
        """
        CREATE TABLE Games (
            game_id VARCHAR(20) PRIMARY KEY,
            game_date DATE,
            team_id_1 INTEGER REFERENCES Teams(team_id),
            team_id_2 INTEGER REFERENCES Teams(team_id),
            team_1 VARCHAR(3),
            team_2 VARCHAR(3),
            pts_1 INTEGER,
            pts_2 INTEGER
        )
        """,
        """
        CREATE TABLE PlayerStats (
            stat_id SERIAL PRIMARY KEY,
            game_id VARCHAR(20) REFERENCES Games(game_id),
            player_id INTEGER REFERENCES Players(player_id),
            points INTEGER,
            rebounds INTEGER,
            assists INTEGER,
            steals INTEGER,
            blocks INTEGER,
            turnovers INTEGER,
            field_goal_percentage DECIMAL(5, 2)
        )
        """,
    )

    try:
        # Connect to the PostgreSQL database
        conn = psycopg2.connect(dbname=DB_NAME, user=DB_USER, password=DB_PASSWORD, host=DB_HOST, port=DB_PORT)
        cur = conn.cursor()

        # Execute each SQL command to create the tables
        for command in commands:
            cur.execute(command)

        # Commit the changes and close the connection
        conn.commit()
        cur.close()
        print("Tables created successfully.")
    except (Exception, psycopg2.DatabaseError) as error:
        print(f"Error: {error}")
    finally:
        if conn is not None:
            conn.close()

# Call the function to create the tables
create_tables()
```

```python
# Insert the data into the PostgreSQL database

def insert_teams(teams_data):
    # SQL statement to insert a team into the Teams table
    insert_team = """
        INSERT INTO Teams (team_id, team_name, abbreviation, city)
        VALUES (%s, %s, %s, %s)
    """

    try:
        # Connect to the PostgreSQL database
        conn = psycopg2.connect(dbname=DB_NAME, user=DB_USER, password=DB_PASSWORD, host=DB_HOST, port=DB_PORT)
        cur = conn.cursor()

        # Insert each team into the Teams table
        for team in teams_data:
            cur.execute(insert_team, (team['team_id'], team['team_name'], team['abbreviation'], team['city']))

        # Commit the changes and close the connection
        conn.commit()
        cur.close()
        print("Teams inserted successfully.")
    except (Exception, psycopg2.DatabaseError) as error:
        print(f"Error: {error}")
    finally:
        if conn is not None:
            conn.close()

            
```

```python
def insert_players(players_data):
    # SQL statement to insert a player into the Players table
    insert_player = """
        INSERT INTO Players (player_id, player_name, is_active)
        VALUES (%s, %s, %s)
    """

    try:
        # Connect to the PostgreSQL database
        conn = psycopg2.connect(dbname=DB_NAME, user=DB_USER, password=DB_PASSWORD, host=DB_HOST, port=DB_PORT)
        cur = conn.cursor()

        # Insert each player into the Players table
        for player in players_data:
            cur.execute(insert_player, (player['player_id'], player['player_name'], player['is_active']))

        # Commit the changes and close the connection
        conn.commit()
        cur.close()
        print("Players inserted successfully.")
    except (Exception, psycopg2.DatabaseError) as error:
        print(f"Error: {error}")
    finally:
        if conn is not None:
            conn.close()
```

```python
def insert_games(games_data):
    # SQL statement to insert a game into the Games table
    insert_game = """
        INSERT INTO Games (game_id, game_date, team_id_1, team_id_2, team_1, team_2, pts_1, pts_2)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
    """

    try:
        # Connect to the PostgreSQL database
        conn = psycopg2.connect(dbname=DB_NAME, user=DB_USER, password=DB_PASSWORD, host=DB_HOST, port=DB_PORT)
        cur = conn.cursor()

        # Insert each game into the Games table
        for game in games_data.keys():
            game_data = games_data[game]
            # print(game_data)
            cur.execute(insert_game, (game, game_data['GAME_DATE'], game_data['TEAM_ID_1'], game_data['TEAM_ID_2'], game_data['TEAM_1'], game_data['TEAM_2'], game_data['PTS_1'], game_data['PTS_2']))
        conn.commit()
        cur.close()
        print("Games inserted successfully.")
    except (Exception, psycopg2.DatabaseError) as error:
        print(f"Error: {error}")
    finally:
        if conn is not None:
            conn.close()
```

```python
def add_player_stats(player_game_stats):
    # SQL statement to insert player stats into the PlayerStats table
    insert_player_stats = """
        INSERT INTO PlayerStats (game_id, player_id, points, rebounds, assists, steals, blocks, turnovers, field_goal_percentage)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
    """

    try:
        # Connect to the PostgreSQL database
        conn = psycopg2.connect(dbname=DB_NAME, user=DB_USER, password=DB_PASSWORD, host=DB_HOST, port=DB_PORT)
        cur = conn.cursor()

        # Insert each player's stats into the PlayerStats table
        for player_stat in player_game_stats:
            cur.execute(insert_player_stats, (player_stat['Game_ID'], player_stat['Player_ID'], player_stat['PTS'], player_stat['REB'], player_stat['AST'], player_stat['STL'], player_stat['BLK'], player_stat['TOV'], player_stat['FG_PCT']))
        conn.commit()
        cur.close()
        print("Player stats inserted successfully.")
    except (Exception, psycopg2.DatabaseError) as error:
        print(f"Error: {error}")
    finally:
        if conn is not None:
            conn.close()
```
